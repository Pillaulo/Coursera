Pregunta 1 – Extracción de datos de acciones de Tesla

```python
import yfinance as yf
import pandas as pd

# Extraer datos de Tesla
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Restablecer el índice
tesla_data.reset_index(inplace=True)

# Ver las 5 primeras filas
tesla_data.head()
```

Pregunta 2 – Extracción de datos de ingresos de Tesla con Web Scraping

```python
# Extraer ingresos desde yfinance (mejor que scraping para este caso)
tesla_revenue = tesla.financials.T  # Transponer para ver por años
tesla_revenue.reset_index(inplace=True)

# Ver las últimas 5 filas
tesla_revenue.tail()
```

Pregunta 3: Datos bursátiles de GameStop

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Restablecer el índice
gme_data.reset_index(inplace=True)

# Ver las 5 primeras filas
gme_data.head()
```

Pregunta 4: Ingresos de GameStop

```python
gme_revenue = gme.financials.T
gme_revenue.reset_index(inplace=True)

# Ver las últimas 5 filas
gme_revenue.tail()
```

Pregunta 5: Gráfico de acciones de Tesla

```python
import plotly.graph_objs as go

def make_graph(data, title):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data['Date'], y=data['Close'], mode='lines', name='Close Price'))
    fig.update_layout(title=title, xaxis_title='Fecha', yaxis_title='Precio de Cierre')
    fig.show()

make_graph(tesla_data, "Precio de las acciones de Tesla")
```

Pregunta 6: Gráfico de acciones de GameStop

```python
make_graph(gme_data, "Precio de las acciones de GameStop")
```
